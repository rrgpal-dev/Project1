## Pennsylvania Data

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

ballots_by_party = "Resources/ballots_by_party.csv"
ballots_by_age = "Resources/ballots_by_age.csv"
party_affiliations = "Resources/party_affiliations.csv"

party_df = pd.read_csv(ballots_by_party)
age_df = pd.read_csv(ballots_by_age)
pa_df = pd.read_csv(party_affiliations)
```

## Early Vote 2016 (USA)
This data includes mail-in ballots and early-voting (available in select states).  
Data provided by the U.S. Election Assistance Commission (eac.gov).

```python
labels = 'Early Vote 2016', 'Election Day Vote 2016'
sizes = [41, 59]
explode = (0, 0.1) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
```
